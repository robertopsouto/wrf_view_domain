## Vendo as informações (meta-dados) do arquivo

```python
import salem
```

```python
%%writefile namelist.wps
&geogrid
 parent_id         = 1,
 parent_grid_ratio = 1,
 i_parent_start    = 1,
 j_parent_start    = 1,
 e_we          = 160,
 e_sn          = 140,
 geog_data_res = '30s',
 dx = 9000,
 dy = 9000,
 map_proj =  'mercator',
 ref_lat   =  -2.43,
 ref_lon   =  -54.72,
 truelat1  =  -2.43,
/
```

```python
from salem import geogrid_simulator
g, maps = geogrid_simulator('namelist.wps', do_maps=True, map_kwargs=None)
```
